```python
# Substituição do traço na coluna floor -> Erro
data[data.floor == '-'].floor = 0
```

```python
# Realizando copia da base
dataCopy = data.copy()
```

```python
# Substituição bem sucedida
dataCopy['floor'] = dataCopy['floor'].replace("-",0)
```

```python
# alterando tipo da coluna
dataCopy['floor'] = dataCopy['floor'].astype(np.int8)
```

```python
dataCopy.head()
```

```python
dataCopy.info()
```

```python
# lista de dados com cifrão
cifrao = ['hoa',	'rent amount',	'property tax',	'fire insurance',	'total']
```

```python
# removendo colunas com cifrão
dados_sem_cifrao = dataCopy.drop(cifrao,axis=1)
```

```python
dados_cifrao = dataCopy.filter(cifrao)
```

```python
dados_cifrao
```

```python
# usando applymap para aplicar uma alteração em todas as colunas 
dados_cifrao_corrigido = dados_cifrao.applymap(lambda x: x.replace('R$','').replace(',','0'))
```

```python
# juntando dois df com concat
dados_ajustados = pd.concat([dados_sem_cifrao,dados_cifrao_corrigido],axis = 1)
```

```python
teste = ['total','fire insurance']
```

```python
dados_ajustados[teste] = dados_ajustados[teste].astype(np.int8)
```

```python
# localizando colunas sem info
for c in cifrao:
    print (dados_ajustados[c].value_counts().head(10))
```

```python
dados_ajustados_copy = dados_ajustados.copy()
dados_ajustados_copy
```

```python
dados_ajustados_copy = dados_ajustados_copy.drop(semvalor1)
dados_ajustados_copy = dados_ajustados_copy.drop(semvalor0)
dados_ajustados_copy = dados_ajustados_copy.drop(semvalor2)
```
